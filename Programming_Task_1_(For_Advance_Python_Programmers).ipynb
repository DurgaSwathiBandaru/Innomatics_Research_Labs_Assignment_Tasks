### 1. Running Sum of 1d Array

```python
# Input: nums = [1,2,3,4]
# Output: [1,3,6,10]

nums = [1,2,3,4]
result = []
total = 0
for i in nums:
  total += i
  result.append(total)
print(result)
```

```python
class Solution():
  def __init__(self,nums):
    self.nums = nums
  def runningSum(self):
    result = []
    total = 0
    for i in self.nums:
      total += i
      result.append(total)
    return result

s1 = Solution([1,2,3,4])
print(s1.runningSum())

s2 = Solution([1,1,1,1,1])
print(s2.runningSum())


s3 = Solution([3,1,2,10,1])
print(s3.runningSum())
```

### 2. Shuffle the Array

```python
''''Input: nums = [2,5,1,3,4,7], n = 3
Output: [2,3,5,4,1,7]
Explanation: Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3,5,4,1,7].'''

class Solution1():
  def __init__(self,nums):
    self.nums = nums
  def shuffle_array(self,n):
    result = []
    x = self.nums[0:n]
    y = self.nums[n:]
    for i in range(n):
      result.append(x[i])
      result.append(y[i])
    return result

a1 = Solution1([2,5,1,3,4,7])
print(a1.shuffle_array(3))

a2 = Solution1([1,2,3,4,4,3,2,1])
print(a2.shuffle_array(4))

a3 = Solution1([1,1,2,2])
print(a3.shuffle_array(2))
```

### 3. Kids With the Greatest Number of Candies

```python
class Greatest_no_candies():
  def __init__(self,nums):
    self.nums = nums
  def kids_with_candies(self,extra_candies):
    max_candies = max(self.nums)
    result = []
    for i in self.nums:
      if (i + extra_candies) >= max_candies:
        result.append(True)
      else:
        result.append(False)
    return result

k = Greatest_no_candies([2,3,5,1,3])
print(k.kids_with_candies(3))

k1 = Greatest_no_candies([4,2,1,1,2])
print(k1.kids_with_candies(1))

k2 = Greatest_no_candies([12,1,12])
print(k2.kids_with_candies(10))
```
